## Step 6. Initial Data Transformation and Feature Engineering
#### Used pandas to clean and transform data
#### Added Family size (sibsp + parch + 1 for self) column
#### Added Is child column.  True if less than 16 years old.
#### Add Fare per person (fare divided by family size)

```python
# Data Transformation and Feature Engineering for Titanic DataFrame

# 1. Check for missing values
print("Missing values per column:\n", titanic_df.isnull().sum())

# 2. Drop rows with missing values in key columns (optional, or you can fill them)
titanic_df = titanic_df.dropna(subset=['age', 'fare', 'embarked'])

# 3. Standardize column names (lowercase, replace spaces with underscores)
titanic_df.columns = titanic_df.columns.str.lower().str.replace(' ', '_')

# 4. Remove duplicate rows (if any)
titanic_df = titanic_df.drop_duplicates()

# 5. Convert categorical columns to category dtype for efficiency
categorical_cols = ['sex', 'class', 'embarked', 'who', 'deck', 'embark_town', 'alive']
for col in categorical_cols:
    if col in titanic_df.columns:
        titanic_df[col] = titanic_df[col].astype('category')

# 6. Feature Engineering: Add new features
# Example: Family size (sibsp + parch + 1 for self)
titanic_df['family_size'] = titanic_df['sibsp'] + titanic_df['parch'] + 1

# Example: Is child (age < 16)
titanic_df['is_child'] = titanic_df['age'] < 16

# Example: Fare per person (fare divided by family size)
titanic_df['fare_per_person'] = titanic_df['fare'] / titanic_df['family_size']

# 7. Rename the 'parch' column to 'parents_children' for clarity and consistency
if 'alive' in titanic_df.columns:
    titanic_df = titanic_df.rename(columns={'parch': 'parents_children'})

# 8. Display cleaned and transformed DataFrame info
titanic_df.info()
titanic_df.head()
```

## Step 7. Initial Visualizations
####   * Used matplotlib to show pairplot's.  
####   * Added a heatmap to show two numerical variable of fare by class and embarkation port.

```python
# Plot average fare by passenger class and embarkation port as a heatmap

# Create a pivot table of average fare by class and embarkation port
fare_pivot = titanic_df.pivot_table(values='fare', index='class', columns='embarked', aggfunc='mean', observed=False)

# Plot the heatmap
sns.heatmap(fare_pivot, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Average Fare by Class and Embarkation Port')
plt.xlabel('Embarked')
plt.ylabel('Passenger Class')
plt.show()
```

####   * Pairplots by aga, fare, siblings and parents / children.  
####   * Stacked bar showing survival by class and sex.
####   * Pie Chart showing proportion of passengers by embarked port.

```python
# Simpler pairplot: Only select a few key numerical columns and color by survival
sns.pairplot(titanic_df, vars=['age', 'fare', 'sibsp', 'parents_children'], hue='survived')
plt.suptitle('Pairplot of Age, Fare, SibSp, and Parents children by Survival', y=1.02)
plt.show()

# To silence the FutureWarning, add observed=False to the groupby call

survival_by_class_sex = (
    titanic_df
    .groupby(['class', 'sex'], observed=False)['survived']
    .value_counts()
    .unstack()
    .fillna(0)
)
survival_by_class_sex.plot(kind='bar', stacked=True)
plt.title('Survival Count by Class and Sex')
plt.xlabel('Class, Sex')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()

# Pie Chart: Proportion of passengers by embarkation port
embarked_counts = titanic_df['embarked'].value_counts()
embarked_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Proportion of Passengers by Embarkation Port')
plt.ylabel('')
plt.show()
```
